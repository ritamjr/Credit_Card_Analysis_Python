# Level 3: Bivariate Analysis (Relationship Between Two Variables)

## 1. Income vs. Spending Behavior

### Problem:
**Does higher income lead to higher credit card spending?**


```python
# Income Buckets
bins = [0, 20000, 40000, 60000, 80000, 100000, 150000, 200000]
labels = ['<20K', '20K-40K', '40K-60K', '60K-80K', '80K-100K', '100K-150K', '150K+']
df['Income_Group'] = pd.cut(df['Income'], bins=bins, labels=labels)


plt.figure(figsize=(8,5))
sns.barplot(x=df['Income_Group'], y=df['Total_Transaction_Amount'], estimator=np.mean, palette='viridis')

plt.title('Income Groups vs. Average Transaction Amount', fontsize=14)
plt.xlabel('Income Group', fontsize=12)
plt.ylabel('Average Transaction Amount', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

##Insights:
### Income Groups vs. Average Transaction Amount **(Bar Chart)**

- Higher income groups tend to spend more on transactions, but there is no strict linear correlation.
- Middle-income groups also show significant transaction amounts, making them an important target.

---



## 2. Utilization Ratio vs. Spending Behavior

### Problem:
Does a higher credit utilization ratio lead to increased spending risks?

## Why This Matters?

High utilization may indicate financial stress, leading to delinquency risks.

Helps banks identify risky customers who may need credit limit adjustments or financial guidance.

 Can be used to create better credit policies and risk management strategies.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Average_Utilization_Ratio'], y=df['Total_Transaction_Amount'], alpha=0.5)

plt.title('Utilization Ratio vs. Spending Behavior', fontsize=14)
plt.xlabel('Average Utilization Ratio', fontsize=12)
plt.ylabel('Total Transaction Amount', fontsize=12)
plt.show()
```

## Insights:
### Utilization Ratio vs. Spending Behavior **(Scatter Plot)**

- Customers with a higher utilization ratio do not necessarily have the highest spending.
- Some high-spending customers maintain a low utilization ratio, likely due to high credit limits.

## 5. Relationship Between Credit Limit and Total Transaction Amount

### Problem:

If no clear trend exists, credit limits might not be the best predictor of transaction amounts.

A flat regression line could mean other factors (income, spending habits) play a bigger role.

High spending, low credit limit customers might be at risk of exceeding their limits or accumulating debt.

Low spending, high credit limit customers could indicate inefficient credit allocation by banks.

## Insights:


### Credit Limit vs. Total Transaction Amount **(Regression Plot)**

- A positive trend exists between credit limit and transaction amount, but with variability.
- Some high-credit-limit customers do not necessarily spend more, while some low-credit-limit customers maximize their spending.

## 8.  Income vs. Utilization Ratio

### Problem:
**Does income level impact credit utilization risk?**

```python
#income bins
bins = [0, 30000, 60000, 100000, 150000, 200000, df['Income'].max()]
labels = ['<30K', '30K-60K', '60K-100K', '100K-150K', '150K-200K', '200K+']


df['Income_Bracket'] = pd.cut(df['Income'], bins=bins, labels=labels)

#average utilization per income bracket
income_utilization = df.groupby('Income_Bracket')['Average_Utilization_Ratio'].mean().reset_index()


plt.figure(figsize=(8,5))
sns.barplot(x='Income_Bracket', y='Average_Utilization_Ratio', data=income_utilization, palette='magma')

plt.title('Income vs. Average Utilization Ratio', fontsize=14)
plt.xlabel('Income', fontsize=12)
plt.ylabel('Average Utilization Ratio', fontsize=12)
plt.show()
```

## Top 5 States and Bottom % States





```python

revenue_by_state = df.groupby('State_Code')['Total_Transaction_Amount'].sum()


revenue_by_state_percentage = (revenue_by_state / revenue_by_state.sum()) * 100


top_5_states = revenue_by_state_percentage.nlargest(5).reset_index()
top_5_states.columns = ['State_Code', 'Grand_Total_%']


print(top_5_states)




plt.figure(figsize=(8, 5))
ax = sns.barplot(data=top_5_states, x='Grand_Total_%', y='State_Code', palette='viridis')


for index, value in enumerate(top_5_states['Grand_Total_%']):
    ax.text(value + 0.5, index, f'{value:.1f}%', va='center', fontsize=12)


plt.xlabel("Grand Total % of Revenue")
plt.ylabel("State Code")
plt.title("Top 5 States by Revenue Percentage")
plt.xlim(0, top_5_states['Grand_Total_%'].max() + 5)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()

```

##Insights
###Top 5 States by Revenue Percentage **(Bar Chart)**
- A few states contribute significantly to total revenue, indicating strong regional performance.
- Expansion strategies could focus on high-revenue states to maximize growth.

---



```python

state_revenue = df.groupby('State_Code')['Total_Transaction_Amount'].sum().reset_index()


grand_total_revenue = state_revenue['Total_Transaction_Amount'].sum()


state_revenue['Grand_Total_%'] = (state_revenue['Total_Transaction_Amount'] / grand_total_revenue) * 100


bottom_5_states = state_revenue.nsmallest(5, 'Grand_Total_%')




plt.figure(figsize=(8, 5))
ax = sns.barplot(data=bottom_5_states, x='Grand_Total_%', y='State_Code', palette='coolwarm')


for index, value in enumerate(bottom_5_states['Grand_Total_%']):
    ax.text(value + 0.2, index, f'{value:.2f}%', va='center', fontsize=12)


plt.xlabel("Grand Total % of Revenue")
plt.ylabel("State Code")
plt.title("Bottom 5 States by Revenue Percentage")
plt.xlim(0, bottom_5_states['Grand_Total_%'].max() + 1)
plt.grid(axis='x', linestyle='--', alpha=0.5)


plt.show()

```

### **Insights:**  
### Bottom 5 States by Revenue Percentage **(Bar Chart)**
- These states contribute very little to overall revenue, suggesting weaker market presence.
- Marketing efforts or service improvements may be needed in these areas.

## Financial Overview by Customer Job: Income, Revenue & Interest Trends

```python



job_summary = df.groupby("Customer_Job").agg({
    "Income": "sum",
    "Total_Transaction_Amount": "sum",
    "Interest_Earned": "sum"
}).reset_index()


job_summary.columns = ["Customer Job", "Sum of Income", "Sum of Revenue", "Sum of Interest Earned"]

empty_row = pd.DataFrame([["", "", "", ""]], columns=job_summary.columns)


grand_total = pd.DataFrame([[
    "Grand Total",
    job_summary["Sum of Income"].sum(),
    job_summary["Sum of Revenue"].sum(),
    job_summary["Sum of Interest Earned"].sum()
]], columns=job_summary.columns)


job_summary = pd.concat([job_summary, empty_row, grand_total], ignore_index=True)


print(job_summary.to_string(index=False))




job_summary_chart = job_summary.iloc[:-2]


fig, ax = plt.subplots(figsize=(10, 6))
job_summary_chart.set_index("Customer Job").plot(kind="barh", stacked=True, ax=ax, colormap="viridis",)


ax.set_title("Financial Summary by Customer Job", fontsize=14, fontweight="bold")
ax.set_xlabel("Amount ($)", fontsize=12)
ax.set_ylabel("Customer Job", fontsize=12)
ax.legend(title="Metric")

plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()



```

## Insights:
### Financial Summary by Customer Job **(Stacked Bar Chart)**
- Different job categories show varying levels of income, revenue contribution, and interest earned.
- Professionals and high-income jobs tend to contribute more, while others contribute less.
